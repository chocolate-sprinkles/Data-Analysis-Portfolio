```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from bs4 import Comment
```

```python
url = "https://www.basketball-reference.com/playoffs/NBA_2023.html"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
comment_ids = ['id="per_poss-team','id="per_poss-opponent','id="shooting-team','id="shooting-opponent']

# for comment in soup(text=lambda text: isinstance(text, Comment)):
#     for comment_id in commented_ids:
#         print(comment_id)
#         if comment_id in comment.string:
#             tag = BeautifulSoup(comment,"html.parser")
#             comment.replace_with(tag)

tags = {}

for comment in soup(text=lambda text: isinstance(text, Comment)):
    for comment_id in comment_ids:
        if comment_id in comment.string:
            tags[comment_id] = BeautifulSoup(comment,"html.parser")
    # if 'id="per_poss-team' in comment.string:
    #     tags['id="per_poss-team'] = BeautifulSoup(comment,"html.parser")

tags
print("Done")
```
