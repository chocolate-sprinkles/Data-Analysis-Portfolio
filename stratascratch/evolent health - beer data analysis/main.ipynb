```python
import pandas as pd
import datetime
import seaborn as sns
```

```python
df = pd.read_csv("BeerDataScienceProject.tar.bz2", compression="bz2")
df
```

The provided compressed file BeerDataScienceProject.tar.bz2 contains data about beers and their reviews. It has the following columns:

    beer_ABV - alcohol by volume
    beer_beerId - beer ID
    beer_brewerId - beer brewer ID
    beer_name - beer name
    beer_style - beer style
    review_appearance - review on the beer's appearance
    review_palette - review on the beer's palette (colours)
    review_overall - overall beer review
    review_taste - review on the beer's taste
    review_profileName - profile name of the reviewer
    review_aroma - review on the beer's aroma
    review_text - the full text of the review
    review_time - timestamp when the review was made


# 1. Rank the top 3 breweries which produce the strongest beers.

```python
# strongest beers
df.sort_values(by="beer_ABV",ascending=False)
```

```python
df.groupby("beer_brewerId").max(numeric_only=True)["beer_ABV"].sort_values(ascending=False).head(3)
```

# 2. Which year did beers enjoy the highest ratings?

```python
# There are multiple ratings and this question is a bit ambiguous. Is it singular highest rating or on average?
# review_overall seems to make the most sense to use
# Singular vs Average
# Ex. Year 1 Ratings -> 10,2,2,2,2,2
#     Year 2 Ratings -> 6,6,6,6,6,6
# Singular Highest Rating -> Year 1
# Average Highest Rating -> Year 2
# Most likely asking for average

# Creating a new column for year value
df["review_year"] = df["review_time"].apply(lambda x: datetime.datetime.fromtimestamp(x).year)
df
```

```python
df.groupby(by="review_year").mean(numeric_only=True)["review_overall"].sort_values(ascending=False)
```

```python
# On average, the year 2000 had the highest ratings for beers
```

# 3. Based on the users' ratings, which factors are important among taste, aroma, appearance, and palette?

```python
df["beer_beerId"].value_counts()
```

```python
# multiple reviews per beer, so aggregate
mean_review = df.groupby("beer_beerId")[["review_appearance","review_taste","review_aroma","review_palette","review_overall"]].mean()
mean_review
```

```python
mean_review.corr()
```

```python
sns.heatmap(mean_review.corr(),annot=True)
```

```python
# Given a correlation coefficient of 0.87, aroma seems to be the most important quality 
```
